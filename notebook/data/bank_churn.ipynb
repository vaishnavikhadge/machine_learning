```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\sanik\\Downloads\\archive (1).zip")
df
```

```python
df.shape
```

```python
df.drop('CustomerId',axis=1,inplace=True)
df.drop('Surname',axis=1,inplace=True)
df.drop('RowNumber',axis=1,inplace=True)
df
```

```python
df.isnull().sum()
```

```python
country=pd.get_dummies(df['Geography']).astype(int)
frame=[df,country]
df=pd.concat(frame,axis=1)
df
```

```python
df.isnull().sum()
```

```python
df.drop('Geography',axis=1,inplace=True)
df
```

```python
contry=pd.get_dummies(df['Card Type']).astype(int)
frame=[df,contry]
df=pd.concat(frame,axis=1)
df
```

```python
df.drop('Card Type',axis=1,inplace=True)
df
```

```python
cols_to_scale=['CreditScore','Balance','EstimatedSalary','Satisfaction Score','Point Earned','Age','Tenure']
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
df[cols_to_scale]=scalar.fit_transform(df[cols_to_scale])
df
```

```python
c=pd.get_dummies(df['Gender']).astype(int)
frame=[df,c]
df=pd.concat(frame,axis=1)
df.drop('Gender',axis=1,inplace=True)
df.drop('NumOfProducts',axis=1,inplace=True)
df
```

```python
#data visualization
```

```python
sns.countplot(x='Age',hue='Tenure',data=df)
```

```python
df.dtypes
```

```python
x=df.drop('IsActiveMember',axis=1)
y=df['IsActiveMember']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
len(x_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(15, input_shape=(19,), activation='relu'),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Assuming x_train and y_train are defined
model.fit(x_train, y_train, epochs=20)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred=model.predict(x_test).reshape(-1)
print(y_pred[:10])
y_pred=np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_test,y_pred))
```

```python
report_dict= classification_report(y_test,y_pred,output_dict=True)
report_dict
```

```python
import mlflow
mlflow.set_experiment("experiment 1")
mlflow.set_tracking_uri("http://127.0.0.1:5000")

with mlflow.start_run():
     mlflow.log_metrics({
        'accuracy' : report_dict['accuracy'],
        'recall_class_0' :report_dict['0']['recall'],
        'recall_class_1':report_dict['1']['recall'],
        'f1_score_macro':report_dict['macro avg']['f1-score']
       })
    
```
